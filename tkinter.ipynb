```python
from tkinter import *
top=Tk()
def hello():
    print('hello!')
menubar=Menu(top)
menubar.add_command(label='hello!',command=hello)
menubar.add_command(label='Quit!',command=top.quit)

# display the menu
top.config(menu=menubar)

top.mainloop()

```

```python
from tkinter import Toplevel ,Button,Tk,Menu
top=Tk()
menubar=Menu(top,tearoff=0)
file=Menu(menubar)
file.add_command(label='new')
file.add_command(label='open')
file.add_command(label='save')
file.add_command(label='save as..')
file.add_command(label='close')

file.add_separator()

file.add_command(label='exit',command=top.quit)

menubar.add_cascade(label='File',menu=file)
edit=Menu(menubar,tearoff=0)
edit.add_command(label='undo')

edit.add_separator()

edit.add_command(label='cut')
edit.add_command(label='copy')
edit.add_command(label='paste')
edit.add_command(label='delete')
edit.add_command(label='select all')

menubar.add_cascade(label='Edit',menu=edit)
help=Menu(menubar,tearoff=0)
help.add_command(label='about')
menubar.add_cascade(label='help',menu=help)

top.config(menu=menubar)
top.mainloop()
```

```python
import tkinter
from tkinter import *

# make a frame of desktop
top=Tk()
top.geometry("600x300")

# ----------------------------------------------------------------------------------------------------------------------------

# make new menubar 

menubar=Menu(top,tearoff=0)

# menubar inside of menubar.

file=Menu(menubar,tearoff=0)
edit=Menu(menubar,tearoff=0)
help=Menu(menubar,tearoff=0)

# added commands insider menubar.
file.add_command(label="add")
file.add_command(label="Open")
file.add_command(label="Save")
file.add_command(label='Save as')
file.add_command(label="Close")
file.add_separator()
file.add_command(label="Exit")

edit.add_command(label="Undo")
edit.add_separator()
edit.add_command(label="Cut")
edit.add_command(label="Copy")
edit.add_command(label="Paste")
edit.add_command(label="Delete")
edit.add_command(label="Select all")

help.add_command(label="About")

# insider menubar connect with main menubar

menubar.add_cascade(label="File",menu=file)
menubar.add_cascade(label="Edit",menu=edit)
menubar.add_cascade(label="Help",menu=help)

# main menubar connect with frame 

top.config(menu=menubar)

# menubar closes here.

# -----------------------------------------------------------------------------------------------------------------------------

# make new form inside of desktop

# getting command and run this function.
def sub():
    tkinter.messagebox.showinfo("Submission","You Have Successfully Submitted Your Form .")
    
    
# we have make labels .   
name=Label(top,text="Name").place(x=10,y=50)
email=Label(top,text="Email").place(x=10,y=90)
password=Label(top,text="Password").place(x=10,y=130)

# make submit button for labels.
submit=Button(top,text="Submit",command=sub).place(x=10,y=170)

# make input box for get data.

e1=Entry(top).place(x=100,y=50)
e2=Entry(top).place(x=100,y=90)
e3=Entry(top).place(x=100,y=130)


# to run the all items inside frame(between top to mainloop).

top.mainloop()
```
